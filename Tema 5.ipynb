```python
#Punctul 1
import pandas as pd
import matplotlib.pyplot as plt
dataset = pd.read_csv("facebook_ads_data.csv")
dataset.head()
dataset['ad_date'] = pd.to_datetime(dataset['ad_date'])
daily_reve = dataset.groupby(dataset['ad_date'].dt.day)['total_value'].sum()
daily_expe = dataset.groupby(dataset['ad_date'].dt.day)['total_spend'].sum()
dataset.groupby(dataset['ad_date'].dt.day)['total_spend'].sum()
Romi = (daily_reve-daily_expe)/daily_expe
#display(Romi) 

# Creați o figură și un ax
fig, ax = plt.subplots()

# Trasați suma totală a cheltuielilor publicitare pe zi
ax.plot(daily_expe.index, daily_expe.values, label='Cheltuieli publicitare', marker='o')

# Adăugați etichete și titluri
ax.set_xlabel('Ziua')
ax.set_ylabel('Suma totală a cheltuielilor publicitare')
ax.set_title('Cheltuieli publicitare zilnice')

# Afișați legenda
ax.legend()

# Afișați graficul
plt.show()

# Creați o nouă figură și un ax
fig, ax = plt.subplots()

# Trasați ROMI-ul zilnic
ax.plot(Romi.index, Romi.values, label='ROMI', marker='o', color='orange')

# Adăugați etichete și titluri
ax.set_xlabel('Ziua')
ax.set_ylabel('ROMI zilnic')
ax.set_title('Return on Marketing Investment (ROMI) zilnic')

# Afișați legenda
ax.legend()

# Afișați graficul
plt.show()
```

```python
#Punctul 2
import pandas as pd
import matplotlib.pyplot as plt

# Încarcă fișierul CSV într-un DataFrame
file_path = 'facebook_ads_data.csv'  # Înlocuiește cu calea către fișierul tău
df = pd.read_csv(file_path)

# Grupăm datele după numele campaniei și calculăm suma totală a cheltuielilor
grouped_data = df.groupby('campaign_name')['total_spend'].sum()

# Creăm un grafic pentru suma totală a cheltuielilor publicitare
fig, ax = plt.subplots()
grouped_data.plot(kind='bar', ax=ax)
ax.set_xlabel('Numele Campaniei')
ax.set_ylabel('Suma Totală a Cheltuielilor Publicitare')
ax.set_title('Suma Totală a Cheltuielilor Publicitare pe Campanie')


# Calculăm ROMI-ul pentru fiecare campanie (exemplu de calcul)
df['romi'] = (df['total_value'] - df['total_spend']) / df['total_spend']
# Grupăm datele după numele campaniei și calculăm ROMI-ul total
grouped_romi = df.groupby('campaign_name')['romi'].sum()

# Creăm un grafic pentru ROMI-ul total al fiecărei campanii
fig, ax = plt.subplots()
grouped_romi.plot(kind='bar', ax=ax)
ax.set_xlabel('Numele Campaniei')
ax.set_ylabel('ROMI Total')
ax.set_title('ROMI Total pe Campanie')

plt.show()
```

```python
#Punctul 3
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('facebook_ads_data.csv')
# Lista de campanii pe care vrei să le includem
campanii_dorite = ['Brand']
#'Brand','Crazy discounts','Discounts','Electronics','Expansion','Hobbies','Lookalike','New items','Promos','Trendy','Wholesale'

# Filtrează datele pentru a include doar campaniile dorite
df_filtrat = df[df['campaign_name'].isin(campanii_dorite)]
df['romi'] = (df['total_value'] - df['total_spend']) / df['total_spend']
grouped_romi = df.groupby('campaign_name')['romi'].sum()

plt.figure(figsize=(10, 6))  # Ajustează dimensiunea graficului
plt.boxplot(df_filtrat['romi'], labels=campanii_dorite)
plt.title('Distribuția ROMI-ului zilnic pe campanii')
plt.xlabel('Nume campanie')
plt.ylabel('ROMI zilnic')
plt.show()
```

```python
#Punctul 5
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# Încarcă datele din fișierul 'facebook_ads_data.csv'
df = pd.read_csv('facebook_ads_data.csv')
numeric_columns = df.select_dtypes(include=np.number)
correlation_matrix = numeric_columns.corr()
#Creeaza heat map-ul
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heat Map al Corelației între Indicatori')
plt.show()
```

```python
#Punctul 6
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Încarcă datele din fișierul 'facebook_ads_data.csv'
df = pd.read_csv('facebook_ads_data.csv')

# Creează un scatter plot cu regresie liniară
sns.lmplot(x='total_spend', y='total_value', data=df)
plt.title('Relația dintre total_spend și total_value')
plt.xlabel('Total Spend')
plt.ylabel('Total Value')
plt.show()
```
