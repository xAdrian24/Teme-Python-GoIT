```python
#Punctul 1
import pandas as pd
import matplotlib.pyplot as plt
dataset = pd.read_csv("facebook_ads_data.csv")
dataset.head()
dataset['ad_date'] = pd.to_datetime(dataset['ad_date'])
daily_reve = dataset.groupby(dataset['ad_date'].dt.day)['total_value'].sum()
daily_expe = dataset.groupby(dataset['ad_date'].dt.day)['total_spend'].sum()
dataset.groupby(dataset['ad_date'].dt.day)['total_spend'].sum()
Romi = (daily_reve-daily_expe)/daily_expe
#display(Romi) 

# Creai o figur i un ax
fig, ax = plt.subplots()

# Trasai suma total a cheltuielilor publicitare pe zi
ax.plot(daily_expe.index, daily_expe.values, label='Cheltuieli publicitare', marker='o')

# Adugai etichete i titluri
ax.set_xlabel('Ziua')
ax.set_ylabel('Suma total a cheltuielilor publicitare')
ax.set_title('Cheltuieli publicitare zilnice')

# Afiai legenda
ax.legend()

# Afiai graficul
plt.show()

# Creai o nou figur i un ax
fig, ax = plt.subplots()

# Trasai ROMI-ul zilnic
ax.plot(Romi.index, Romi.values, label='ROMI', marker='o', color='orange')

# Adugai etichete i titluri
ax.set_xlabel('Ziua')
ax.set_ylabel('ROMI zilnic')
ax.set_title('Return on Marketing Investment (ROMI) zilnic')

# Afiai legenda
ax.legend()

# Afiai graficul
plt.show()
```

```python
#Punctul 2
import pandas as pd
import matplotlib.pyplot as plt

# ncarc fiierul CSV ntr-un DataFrame
file_path = 'facebook_ads_data.csv'  # nlocuiete cu calea ctre fiierul tu
df = pd.read_csv(file_path)

# Grupm datele dup numele campaniei i calculm suma total a cheltuielilor
grouped_data = df.groupby('campaign_name')['total_spend'].sum()

# Crem un grafic pentru suma total a cheltuielilor publicitare
fig, ax = plt.subplots()
grouped_data.plot(kind='bar', ax=ax)
ax.set_xlabel('Numele Campaniei')
ax.set_ylabel('Suma Total a Cheltuielilor Publicitare')
ax.set_title('Suma Total a Cheltuielilor Publicitare pe Campanie')


# Calculm ROMI-ul pentru fiecare campanie (exemplu de calcul)
df['romi'] = (df['total_value'] - df['total_spend']) / df['total_spend']
# Grupm datele dup numele campaniei i calculm ROMI-ul total
grouped_romi = df.groupby('campaign_name')['romi'].sum()

# Crem un grafic pentru ROMI-ul total al fiecrei campanii
fig, ax = plt.subplots()
grouped_romi.plot(kind='bar', ax=ax)
ax.set_xlabel('Numele Campaniei')
ax.set_ylabel('ROMI Total')
ax.set_title('ROMI Total pe Campanie')

plt.show()
```

```python
#Punctul 3
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('facebook_ads_data.csv')
# Lista de campanii pe care vrei s le includem
campanii_dorite = ['Brand']
#'Brand','Crazy discounts','Discounts','Electronics','Expansion','Hobbies','Lookalike','New items','Promos','Trendy','Wholesale'

# Filtreaz datele pentru a include doar campaniile dorite
df_filtrat = df[df['campaign_name'].isin(campanii_dorite)]
df['romi'] = (df['total_value'] - df['total_spend']) / df['total_spend']
grouped_romi = df.groupby('campaign_name')['romi'].sum()

plt.figure(figsize=(10, 6))  # Ajusteaz dimensiunea graficului
plt.boxplot(df_filtrat['romi'], labels=campanii_dorite)
plt.title('Distribuia ROMI-ului zilnic pe campanii')
plt.xlabel('Nume campanie')
plt.ylabel('ROMI zilnic')
plt.show()
```

```python
#Punctul 5
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# ncarc datele din fiierul 'facebook_ads_data.csv'
df = pd.read_csv('facebook_ads_data.csv')
numeric_columns = df.select_dtypes(include=np.number)
correlation_matrix = numeric_columns.corr()
#Creeaza heat map-ul
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heat Map al Corelaiei ntre Indicatori')
plt.show()
```

```python
#Punctul 6
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# ncarc datele din fiierul 'facebook_ads_data.csv'
df = pd.read_csv('facebook_ads_data.csv')

# Creeaz un scatter plot cu regresie liniar
sns.lmplot(x='total_spend', y='total_value', data=df)
plt.title('Relaia dintre total_spend i total_value')
plt.xlabel('Total Spend')
plt.ylabel('Total Value')
plt.show()
```
